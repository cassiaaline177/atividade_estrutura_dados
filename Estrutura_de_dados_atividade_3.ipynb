## 1) Lista Ligada Circular

Uma lista ligada circular é uma estrutura em que o último nó aponta novamente para o primeiro, formando um ciclo.

a) A referência do último nó aponta para o primeiro nó da lista, não sendo nula.

b) Aplicação prática: sistemas de escalonamento circular, como controle de processos em sistemas operacionais ou playlists em modo loop.

c) Deve-se ter cuidado ao percorrer a lista para não entrar em loop infinito. É necessário definir uma condição de parada, como voltar ao primeiro nó.

## 2) Lista Duplamente Encadeada

É uma lista em que cada nó possui referências para o próximo e para o anterior.

a) Atributos:
- dado
- proximo
- anterior

b) Vantagem: permite navegação nos dois sentidos (ida e volta).

c) Impacto: maior consumo de memória, pois cada nó armazena duas referências.

```python
# Questão 3
class NoDuplo:
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None
        self.anterior = None
```

## 4) Operações em Lista Ligada Simples

a) Inserção no início: O(1)  
b) Inserção no final: O(n)  
c) Busca por elemento: O(n)  
d) Acesso por posição: O(n)

Isso ocorre porque a lista ligada não possui acesso direto por índice, exigindo percurso sequencial.

```python
# Questão 5
class No:
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None

class ListaLigada:
    def __init__(self):
        self.inicio = None

    def adicionar_inicio(self, valor):
        novo = No(valor)
        novo.proximo = self.inicio
        self.inicio = novo
```

A complexidade é O(1) porque a inserção ocorre diretamente no início, sem necessidade de percorrer a lista.

## 6) Inserção no Final

a) Complexidade: O(n), pois é necessário percorrer toda a lista.

b) Pode ser reduzida mantendo um ponteiro para o último nó.

c) Implementação abaixo.

```python
def adicionar_fim(self, valor):
    novo = No(valor)

    if not self.inicio:
        self.inicio = novo
        return

    atual = self.inicio
    while atual.proximo:
        atual = atual.proximo

    atual.proximo = novo
```

```python
# Questão 7

def percorrer(self):
    atual = self.inicio
    while atual:
        print(atual.dado)
        atual = atual.proximo
```

a) O percurso começa pelo primeiro nó porque ele é o único ponto de acesso à lista.

b) A condição de parada ocorre quando o ponteiro chega em None, indicando o final da lista.

```python
# Questão 8

def inserir_posicao(self, valor, pos):
    if pos == 0:
        self.adicionar_inicio(valor)
        return

    atual = self.inicio
    for _ in range(pos - 1):
        atual = atual.proximo

    novo = No(valor)
    novo.proximo = atual.proximo
    atual.proximo = novo
```

a) Se a posição for 0, o elemento é inserido no início.

b) Complexidade: O(n), pois pode ser necessário percorrer a lista.

c) Deve-se verificar se a posição é válida para evitar acessar nós inexistentes.

```python
# Questão 9

def obter_elemento(self, pos):
    atual = self.inicio
    for _ in range(pos):
        atual = atual.proximo
    return atual.dado
```

A complexidade é O(n) porque é necessário percorrer a lista desde o início até a posição desejada, já que não há acesso direto por índice.
