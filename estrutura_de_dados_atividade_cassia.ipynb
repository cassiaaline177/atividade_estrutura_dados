## 1)  Explique o funcionamento da busca linear e indique

A busca linear percorre a lista elemento por elemento até encontrar o valor desejado.

a) Deve ser utilizada em listas não ordenadas ou pequenas.

b) Complexidade no pior caso: O(n)

c) Vantagem: simples e funciona em qualquer lista.
Desvantagem: lenta para grandes volumes de dados.

## 2) Implementação da Busca Linear

Função que retorna o índice do elemento se encontrado, ou -1 caso contrário.

```python
def busca_linear(lista, valor):
    for i in range(len(lista)):
        if lista[i] == valor:
            return i
    return -1

print(busca_linear([10, 20, 30, 40], 30))
```

## 3) Busca Binária

Lista: [5, 12, 18, 27, 34, 42, 50]

a) Passo a passo para encontrar 34:

1. Meio = 27 → 34 é maior → busca na direita  
2. Meio = 42 → 34 é menor → busca na esquerda  
3. Meio = 34 → encontrado

b) Complexidade: O(log n)

c) A lista precisa estar ordenada para que seja possível descartar metade dos elementos a cada comparação.

## 4) Busca por Interpolação

A busca por interpolação estima a posição do elemento com base no valor procurado.

a) Mais eficiente em dados uniformemente distribuídos.

b) Complexidade média: O(log log n), podendo ser mais rápida que a busca binária.

c) Pode ter desempenho ruim quando os dados não são uniformes, chegando a O(n) no pior caso.

## 5) Selection Sort

Lista: [29, 10, 14, 37, 13]

a) Primeiras iterações:

Iteração 1: menor = 10 → troca com 29  
[10, 29, 14, 37, 13]

Iteração 2: menor = 13 → troca com 29  
[10, 13, 14, 37, 29]

b) Complexidade: O(n²)

c) É considerado elementar porque usa comparações simples e não utiliza técnicas avançadas como divisão e conquista.

## 6) Insertion Sort

Algoritmo que insere cada elemento na posição correta dentro da parte já ordenada da lista.

```python
def insertion_sort(lista):
    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1

        while j >= 0 and lista[j] > chave:
            lista[j + 1] = lista[j]
            j -= 1

        lista[j + 1] = chave
    return lista

print(insertion_sort([29, 10, 14, 37, 13]))
```

O insertion sort apresenta bom desempenho em listas pequenas ou quase ordenadas, pois realiza poucas trocas nesses casos.

## 7) Quicksort vs Mergesort

a) Complexidade média do Quicksort: O(n log n)

b) Complexidade garantida do Mergesort: O(n log n)

c) O Mergesort utiliza divisão e conquista de forma mais estruturada, pois sempre divide a lista ao meio.

d) O Mergesort necessita de memória auxiliar adicional, pois cria listas temporárias durante a intercalação.

## 8) Implementação do Quicksort

```python
def quicksort(lista):
    if len(lista) <= 1:
        return lista

    pivo = lista[0]
    menores = [x for x in lista[1:] if x <= pivo]
    maiores = [x for x in lista[1:] if x > pivo]

    return quicksort(menores) + [pivo] + quicksort(maiores)

lista = [45, 23, 89, 12, 78, 34, 56]
print(quicksort(lista))
```

c) Diferença:
